# YouTube Video Analysis using Python

```python
import numpy as np
import pandas as pd
```

### Reading CSV Files

```python
youtube_df=pd.read_csv("Youtubedataset.csv")
```

### Cleaning Youtube_df Data Frame

#checking all the columns
youtube_df.columns

```python
#dropping columns unnamed: 17 and 18
youtube_df.drop(['Unnamed: 17','Unnamed: 18'],axis=1,inplace=True)
```

```python
# checking df for the updates
youtube_df.columns
```

```python
#checking data types of eanch and every column
youtube_df.info()
```

```python
#checking unique values in category_id column
youtube_df.category_id.unique()
```

```python
#checking unique values in Tag_count column
youtube_df.Tag_count.unique()
```

```python
#checking unique values in comment_count  column
youtube_df.comment_count .unique()
```

```python
#checking unique values in likes column
youtube_df.likes.unique()
```

```python
#checking unique values in dislike column
youtube_df.dislike.unique()
```

```python
#checking unique values in views column
youtube_df.views.unique()
```

```python
# cleaning category_id column
youtube_df.category_id.str.replace('"','')
youtube_df['category_id'] = pd.to_numeric(youtube_df.category_id, errors='coerce')
```

```python
# cleaning Tag_count column
youtube_df.Tag_count.str.replace('','0').replace('alissa ashley|alissa ashley makeup|hooded eye makeup|makeup for hooded eyes','0')
youtube_df['Tag_count'] = pd.to_numeric(youtube_df.Tag_count, errors='coerce')
```

```python
# cleaning comment_count column
youtube_df.comment_count.str.replace('','0')
youtube_df['comment_count'] = pd.to_numeric(youtube_df.comment_count, errors='coerce')
```

```python
# cleaning likes column
youtube_df.likes.str.replace('','0')
youtube_df['likes'] = pd.to_numeric(youtube_df.likes, errors='coerce')
```

```python
# cleaning Trend_tag_count column
youtube_df.Trend_tag_count.str.replace('','0')
youtube_df['Trend_tag_count'] = pd.to_numeric(youtube_df.Trend_tag_count, errors='coerce')
```

```python
# cleaning dislike column
youtube_df.dislike.str.replace('','0')
youtube_df['dislike'] = pd.to_numeric(youtube_df.dislike, errors='coerce')
```

```python
# cleaning views column
youtube_df.views.str.replace('','0')
youtube_df['views'] = pd.to_numeric(youtube_df.views, errors='coerce')
```

```python
# converting true to false and false to true in comments_disabled column
def fun(val):
    if val=='FALSE':
        return True
    return False
youtube_df['comment_disabled']=youtube_df['comment_disabled'].apply(fun)
```

```python
def fun2(val):
    if val=='TRUE':
        return True
    return False
youtube_df['like dislike disabled']=youtube_df['like dislike disabled'].apply(fun2)
```

```python
youtube_df['tag appered in title']=youtube_df['tag appered in title'].astype('bool')
```

```python
youtube_df.info()
```

```python
youtube_df['category_id'] = youtube_df['category_id'].fillna(0)
```

```python
youtube_df['category_id']=youtube_df['category_id'].astype('int64')
```

```python
youtube_df=youtube_df.fillna({'subscriber':0,'Trend_day_count':0,'Tag_count':0,'Trend_tag_count':0,'comment_count':0,'likes':0,'dislike':0,'views':0})
youtube_df=youtube_df.astype({'subscriber':'int64','Trend_day_count':'int64','Tag_count':'int64','Trend_tag_count':'int64','comment_count':'int64','likes':'int64','dislike':'int64','views':'int64'})
```

# Asking and Answering Questions

### Q1: What is the total subscriber count for each video category?

```python
youtube_df.category_id.unique()
```

```python
youtube_df.subscriber
```

```python
cat_sub=youtube_df.groupby("category_id")[["subscriber"]].sum()
```

```python
plt.figure(figsize=(12,6))
plt.xlabel("Category Id")
plt.ylabel("Total count of subscriber/1e9")
plt.title("What is the total subscriber count for each video category?")
plt.plot(cat_sub.index,cat_sub/1e9,"b--o",linewidth=3,markersize=10);
```

### Q2: What is the total like/dislike count for each video category?

```python
cat_ld=youtube_df.groupby("category_id")[["likes","dislike"]].sum()
```

```python
n = len(cat_ld)
likes = cat_ld.likes/1e6
dislike = cat_ld.dislike/1e6

fig, ax = plt.subplots()
index = np.arange(n)
bar_width = 0.35
opacity = 0.8

like1 = plt.bar(index, likes, bar_width,
alpha=opacity,
color='b',
label='Likes')

dilike1 = plt.bar(index + bar_width, dislike, bar_width,
alpha=opacity,
color='g',
label='Dislikes')

plt.xlabel('Category')
plt.ylabel('Count (in million)')
plt.title('Like/Dislike Count by Category')
plt.xticks(index + bar_width, cat_ld.index)
plt.legend()
plt.tight_layout()
plt.show();
```

### Q3: List the top 15 videos with highest comment count.
